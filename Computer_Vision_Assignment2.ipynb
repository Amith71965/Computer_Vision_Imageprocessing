```python
#importing Necessary Libraries which is required
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the grayscale image
img = cv2.imread('grayscale_image.jpeg', 0)

# Calculate the histogram of the image
hist, bins = np.histogram(img.flatten(), 256, [0, 256])

# Calculate the cumulative sum of the histogram
cum_sum = hist.cumsum()
```

```python
# Normalizing the cumulative sum to get the lookup table
lookup_table = (cum_sum - cum_sum.min()) * 255 / (cum_sum.max() - cum_sum.min())

# Using the lookup table to perform histogram equalization
equalized_img = np.interp(img.flatten(), bins[:-1], lookup_table).astype(np.uint8)

# Reshaping the equalized image to its original shape
equalized_img = equalized_img.reshape(img.shape)
```

```python
# Now Ploting the original image and its histogram
plt.subplot(221)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.xticks([])
plt.yticks([])
plt.subplot(222)
plt.hist(img.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.title('Histogram of Original Image')
```

The Plot is matching with the plot of Edited software which i used while editing this coloured image

```python
# Now Ploting the equalized image and its histogram
plt.subplot(223)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')
plt.xticks([])
plt.yticks([])
plt.subplot(224)
plt.hist(equalized_img.flatten(), 256, [0, 256], color='r')
plt.xlim([0, 256])
plt.title('Histogram of Equalized Image')
```

```python
# Display the plots
plt.show()
```

Here, I have used OpenCV, NumPy, and Matplotlib libraries to perform histogram equalization and plot the original and equalized images along with their histograms.
